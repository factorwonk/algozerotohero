#### Price Series Transformations

```python
aapl = close.AAPL.copy().to_frame()
print(aapl)
```

```python
# shift function - default is one
aapl.shift(periods=1)
```

```python
aapl["lag1"]=aapl.shift(periods=1)
aapl
```

```python
# Apple close price minus lag close price
aapl.AAPL.sub(aapl.lag1)
```

```python
aapl["Diff1"] = aapl.AAPL.sub(aapl.lag1)
aapl
```

```python
# Percentage Change
aapl.Diff1.div(aapl.lag1)
```

```python
aapl["% Change"] = aapl.Diff1.div(aapl.lag1).mul(100)
aapl
```

```python
# Diff and Diff2 are the same
aapl["Diff2"] = aapl.AAPL.diff(periods=1)
aapl
```

```python
aapl["% Change 2"] = aapl.AAPL.pct_change(periods=1).mul(100)
aapl
```

```python
# Delete columns
del aapl["Diff2"]
aapl
```

```python
aapl.drop(["lag1", "Diff1", "% Change 2"], axis=1, inplace=True)
aapl
```

```python
aapl.rename(columns={'% Change':'Change'}, inplace=True)
aapl
```

#### Resampling

```python
# monthly resample
aapl.AAPL.resample("M").last()
```

```python
# Resample to last business day of month
aapl.AAPL.resample("BM").last()
```

```python
# Calculate Pct Change
aapl.AAPL.resample("BM").last().pct_change(periods=1).mul(100)
```

#### Calculate Mean Return and Standard Dev

```python
aapl.drop("Change", axis=1, inplace=True)
aapl
```

```python
ret = aapl.pct_change().dropna()
ret
```

```python
ret.info()
```

```python
ret.plot(kind="hist", figsize=(12,8), bins=100)
plt.show()
```

#### Metrics

```python
daily_mean_ret = ret.mean()
daily_mean_ret
```

```python
var_daily = ret.var()
var_daily
```

```python
ret.std()
```

```python
stdev_daily = np.sqrt(var_daily)
stdev_daily
```

```python
# Annualised return
annual_mean_ret = daily_mean_ret*252
annual_mean_ret
```

```python
annual_var_return = var_daily * 252
annual_var_return
```

```python
annual_std_return = np.sqrt(annual_var_return)
annual_std_return
```

```python
# Check the above by multiplying std of daily returns with sqrt(252)
ret.std()*np.sqrt(252)
```

```python
more_tickers = ["SPY", "AAPL", "KO", "IBM", "DIS", "MSFT"]
```

```python
more_stocks = yf.download(more_tickers, start="2010-01-01", end="2021-01-01")
```

```python
close = more_stocks.loc[:,"Close"].copy()
```

```python
normclose = close.div(close.iloc[0]).mul(100)
```

```python
normclose.plot(figsize=(15,8), fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
ret = close.pct_change().dropna()
ret.head()
```

```python
ret.describe()
```

```python
# Transpose the above
ret.describe().T
```

```python
# Just grab two columns
summary = ret.describe().T.loc[:,["mean", "std"]]
summary
```

```python
# Annualize figures and annotate a plot
summary_ann = summary.copy()
summary_ann["mean"] = summary_ann["mean"]*252
summary_ann["std"] = summary_ann["std"]*np.sqrt(252)
summary_ann
```

```python
summary_ann.plot.scatter(x="std", y="mean", figsize=(12,8), s=50, fontsize=15)
plt.show()
```

```python
# Add annotations
summary_ann.plot.scatter(x="std", y="mean", figsize=(12,8), s=50, fontsize=15)
for i in summary_ann.index:
    plt.annotate(i, 
    xy=(summary_ann.loc[i,"std"]+0.002, summary_ann.loc[i,"mean"]+0.002), 
    size=15)
plt.xlabel("Annual Risk (std)", fontsize=15)
plt.ylabel("Annual Return", fontsize=15)
plt.title("Equity Risk/Return Trade-off", fontsize=25)
```
